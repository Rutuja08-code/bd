```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/heart - heart.csv')
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop(columns=['stn_code', 'agency', 'sampling_date', 'location_monitoring_station', 'pm2_5', 'state', 'location', 'type', 'date'], errors='ignore')
```

```python
#heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='YlGnBu')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# . Line Plot for 'age'
plt.figure(figsize=(10, 4))
plt.plot(df['age'][:100], color='orange')
plt.title("Age Distribution")
plt.xlabel("Sample Row")
plt.ylabel("Age")
plt.show()
```

```python
#boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='chol', data=df)
plt.title("Boxplot of Cholesterol by Sex")
plt.show()
```

```python
#. Histogram for Age Distribution
plt.figure(figsize=(8, 4))
plt.hist(df['age'], bins=30, color='purple')
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#. Scatter Plot for Age vs Thalachh (Maximum Heart Rate)
plt.figure(figsize=(8, 5))
plt.scatter(df['age'], df['thalachh'], color='teal')
plt.title("Age vs Maximum Heart Rate")
plt.xlabel("Age")
plt.ylabel("Maximum Heart Rate (thalachh)")
plt.show()
```
