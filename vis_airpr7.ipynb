```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Load the dataset
df = pd.read_csv('/content/airquality_data - airquality_data.csv')
```

```python
#cleaning
# Step 3: Drop unnecessary columns
df = df.drop(['stn_code', 'agency', 'sampling_date', 'location_monitoring_station', 'pm2_5'], axis=1, errors='ignore')
# Step 4: Remove missing values
df = df.dropna()

```

```python
#heatmap
plt.figure(figsize=(8, 6))
corr = df[['so2', 'no2', 'rspm', 'spm']].corr()  # Calculate correlation matrix
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)
plt.title("Correlation Heatmap for Air Quality Features")
plt.show()
```

```python
#lineplot

plt.figure(figsize=(10, 4))
plt.plot(df['so2'][:100], label='SO2', color='blue')
plt.plot(df['no2'][:100], label='NO2', color='green')
plt.title("SO2 and NO2 Levels")
plt.xlabel("Sample Row")
plt.ylabel("Value")
plt.legend()
plt.show()

```

```python
#lineplot RSPM level
plt.figure(figsize=(8, 4))
plt.plot(df['rspm'][:100], color='orange')
plt.title("RSPM Levels")
plt.xlabel("Sample Row")
plt.ylabel("RSPM")
plt.show()
```

```python
#boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df[['so2', 'no2', 'rspm', 'spm']])
plt.title("Boxplot of Pollutants")
plt.show()
```

```python
#histogram
plt.figure(figsize=(8, 4))
plt.hist(df['so2'], bins=30, color='purple')
plt.title("Distribution of SO2")
plt.xlabel("SO2 Level")
plt.ylabel("Frequency")
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(df['no2'], df['rspm'], color='teal')
plt.title("NO2 vs RSPM")
plt.xlabel("NO2")
plt.ylabel("RSPM")
plt.show()
```
